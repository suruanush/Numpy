# NumPy
- Short for Numerical Python
- Foundational Package for Numerical computing in python
- Provides an ndarray object which is an efficient multidimensional array providing fast array oriented operations
- Provides a lot of functions

```python
import numpy as np
```

```python
import time
```

## Calculating the square of each element

```python
# using lists
arr_list = [i for i in range(1,10000000,1)]
```

```python
arr_list[:10]
```

```python
start_time = time.time()
sq_arr = [x**2 for x in arr_list]
end_time = time.time()

print("time taken using lists", end_time - start_time)
```

```python
sq_arr[:10]
```

```python
# using Numpy array
arr_numpy = np.arange(1,10000000,1)
```

```python
arr_numpy[:10]
```

```python
start_time = time.time()
sq_numpy = arr_numpy**2
end_time = time.time()

print("time taken using numpy", end_time - start_time)
```
